# Correct Purpose of the Confidence Interval

**2. To estimate the true average pressure of the pipeline and give ourselves some room for error in the estimate.**
&nbsp;<br>
&nbsp;


## Reason

- A confidence interval is used to estimate a population parameter (in this case, the true average pressure $μ$) based on sample data.
- The 95% confidence interval provides a range of values within which we are 95% confident that the true average pressure &μ& lies.
- It accounts for sampling variability and provides a margin of error around the sample mean to estimate the true population mean.

&nbsp;

The other options are incorrect because:

1. The confidence interval is not about estimating the average of the sample measurements (we already have the sample mean).

2. The confidence interval does not predict where individual measurements fall.

3. The confidence interval does not predict where future measurements will fall; it estimates the true population parameter.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

sns.set(style="whitegrid")

np.random.seed(42) 
true_mean_pressure = 100  # True average pressure (unknown)
std_dev = 5               # True standard deviation (unknown)
n = 100  
pressure_measurements = np.random.normal(true_mean_pressure, std_dev, n)

sample_mean = np.mean(pressure_measurements)
sample_std = np.std(pressure_measurements, ddof=1)
standard_error = sample_std / np.sqrt(n)

confidence_level = 0.95
degrees_freedom = n - 1 
confidence_interval = stats.t.interval(confidence_level, degrees_freedom, loc=sample_mean, scale=standard_error)

print(f"Sample Mean Pressure: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")
print(f"95% Confidence Interval for True Mean Pressure: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})")

plt.figure(figsize=(10, 6))
sns.histplot(pressure_measurements, kde=True, color="blue", bins=15, label="Pressure Measurements")
plt.axvline(sample_mean, color="red", linestyle="--", label=f"Sample Mean: {sample_mean:.2f}")
plt.axvline(confidence_interval[0], color="green", linestyle=":", label=f"95% CI Lower Bound: {confidence_interval[0]:.2f}")
plt.axvline(confidence_interval[1], color="green", linestyle=":", label=f"95% CI Upper Bound: {confidence_interval[1]:.2f}")
plt.title("Distribution of Pressure Measurements with 95% Confidence Interval", fontsize=14)
plt.xlabel("Pressure", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.legend()
sns.despine()
plt.show()
```
